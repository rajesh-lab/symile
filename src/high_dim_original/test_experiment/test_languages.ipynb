```python
from collections import Counter
import json
import pandas as pd
```

```python
pt = "/gpfs/scratch/as16583/symile/src/high_dim/data_reference.json"
l_dict = json.load(open(pt))
```

```python
forbidden_list = ["ru", "sv", "pt"]
```

```python
# Remove 'synset_id' and languages in the forbidden_list
for class_name, class_dict in l_dict.items():
    # Remove 'synset_id'
    class_dict.pop('synset_id', None)

    # Remove languages in the forbidden_list
    for lang in forbidden_list:
        class_dict.pop(lang, None)
```

```python
# Create a Counter for each language's values
language_counters = {lang: Counter() for lang in next(iter(l_dict.values())).keys()}

# Count unique values for each language across all class_name entries
for class_name, class_dict in l_dict.items():
    for lang, value in class_dict.items():
        language_counters[lang][value] += 1

# Calculate the overlap between languages
language_overlap = {lang1: {lang2: len(set(language_counters[lang1]) & set(language_counters[lang2]))
                            for lang2 in language_counters if lang1 != lang2}
                    for lang1 in language_counters}

# Print or analyze the overlap
for lang1, overlap_dict in language_overlap.items():
    print(f"Overlap for {lang1}: {overlap_dict}")
```
