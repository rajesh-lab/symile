```python
from collections import Counter
import json
import pandas as pd
```

```python
pt = "/gpfs/scratch/as16583/symile/src/high_dim/data_reference.json"
l_dict = json.load(open(pt))
```

```python
forbidden_list = ["ru", "sv", "pt", "fr", "it", "nl", "id", "zh-CN", "tr", "de", "es", "vi", "pl", "no", "fi", "cs", "cy", "is", "hu",
"he", "hi", "el", "ur", "uk", "th", "te", "ko", "ja"]
```

```python
# Remove 'synset_id' and languages in the forbidden_list
for class_name, class_dict in l_dict.items():
    # Remove 'synset_id'
    class_dict.pop('synset_id', None)

    # Remove languages in the forbidden_list
    for lang in forbidden_list:
        class_dict.pop(lang, None)
```

```python
# Create a Counter for each language's values
language_counters = {lang: Counter() for lang in next(iter(l_dict.values())).keys()}

# Count unique values for each language across all class_name entries
for class_name, class_dict in l_dict.items():
    for lang, value in class_dict.items():
        language_counters[lang][value] += 1

# Calculate the overlap between languages
language_overlap = {lang1: {lang2: len(set(language_counters[lang1]) & set(language_counters[lang2]))
                            for lang2 in language_counters if lang1 != lang2}
                    for lang1 in language_counters}

# Print or analyze the overlap
for lang1, overlap_dict in language_overlap.items():
    print(f"Overlap for {lang1}: {overlap_dict}")
```

```python
class_name_overlaps = Counter()

# Identify class names with overlaps and associated languages
for lang1, overlap_dict in language_overlap.items():
    for lang2, count in overlap_dict.items():
        if count > 0:
            class_name_overlaps[(lang1, lang2)] += 1

# You can also print the detailed overlap information
for (lang1, lang2), count in class_name_overlaps.items():
    print(f"\nOverlap details between {lang1} and {lang2} ({count} class names):")
    for class_name, class_dict in l_dict.items():
        values_lang1 = set(value for lang, value in class_dict.items() if lang == lang1)
        values_lang2 = set(value for lang, value in class_dict.items() if lang == lang2)
        common_values = values_lang1 & values_lang2
        if common_values:
            print(f"Class Name: {class_name}, Common Values: {common_values}")
```

```python
unique_dict = {}
# for each category, how many unique values are there?
for c, v in l_dict.items():
    # Flatten the values across all languages
    all_values = [value for lang, value in v.items() if lang != 'synset_id']

    # Count the unique values
    unique_value_count = len(set(all_values))

    unique_dict[c] = unique_value_count

count = 0
for k, v in unique_dict.items():
    if v == 20:
        count += 1
```
