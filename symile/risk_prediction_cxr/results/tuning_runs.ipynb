```python
symile_df = df[df["loss_fn"] == "symile"]
clip_df = df[df["loss_fn"] == "clip"]
```

```python
# symile_df.groupby(["learning_rate", "weight_decay"]).max()
# group by learning rate and weight decay, and return the full row that corresponds to min over val loss
# Group by 'learning_rate' and 'weight_decay', then find the index of the min 'val_loss'
idx = df.groupby(['loss_fn'])['val_acc_at_1'].idxmax()

# Retrieve the rows corresponding to these indices
result_df = df.loc[idx]
result_df[['loss_fn','pretrained','learning_rate', 'weight_decay', 'val_loss', 'val_acc_at_1', 'val_acc_at_5', 'val_acc_at_10', 'ckpt_pt']]
```
