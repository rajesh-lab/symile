```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import plotly.graph_objects as go
```

```python
save_dir = "/gpfs/scratch/as16583/symile/symile/high_dim/results/umap"
```

```python
#2D UMAP
loss_fn = "symile"
metric = "correlation"

e_a = np.load(f"{save_dir}/{loss_fn}/r_a_2d_{metric}.npy")
e_i = np.load(f"{save_dir}/{loss_fn}/r_i_2d_{metric}.npy")
e_t = np.load(f"{save_dir}/{loss_fn}/r_t_2d_{metric}.npy")

plt.figure(figsize=(6,4))
plt.scatter(e_a[:, 0], e_a[:, 1], color='red', alpha=0.5, label='r_a')
plt.scatter(e_i[:, 0], e_i[:, 1], color='blue', alpha=0.5, label='r_i')
plt.scatter(e_t[:, 0], e_t[:, 1], color='green', alpha=0.5, label='r_t')

# Optionally connect triples with lines
# for j in range(e_a.shape[0]):  # assuming all are the same length
#     # Repeat the first point at the end to close the loop
#     plt.plot([e_a[j, 0], e_i[j, 0], e_t[j, 0], e_a[j, 0]], [e_a[j, 1], e_i[j, 1], e_t[j, 1], e_a[j, 1]], 'k-', alpha=0.1)

plt.title(f'{loss_fn} UMAP, metric = {metric}')
plt.legend()
plt.show()
```

```python
# 3D UMAP

loss_fn = "symile"
metric = "correlation"

e_a = np.load(f"{save_dir}/{loss_fn}/r_a_3d_{metric}.npy")
e_i = np.load(f"{save_dir}/{loss_fn}/r_i_3d_{metric}.npy")
e_t = np.load(f"{save_dir}/{loss_fn}/r_t_3d_{metric}.npy")

# Create a 3D scatter plot using Plotly Express
fig = go.Figure()

# Adding scatter plots for each set of embeddings
fig.add_trace(go.Scatter3d(
    x=e_a[:, 0], y=e_a[:, 1], z=e_a[:, 2],
    mode='markers',
    marker=dict(size=5, color='red', opacity=0.5),
    name='r_a'
))
fig.add_trace(go.Scatter3d(
    x=e_i[:, 0], y=e_i[:, 1], z=e_i[:, 2],
    mode='markers',
    marker=dict(size=5, color='blue', opacity=0.5),
    name='r_i'
))
fig.add_trace(go.Scatter3d(
    x=e_t[:, 0], y=e_t[:, 1], z=e_t[:, 2],
    mode='markers',
    marker=dict(size=5, color='green', opacity=0.5),
    name='r_t'
))

# Connect triples with lines
# for j in range(e_a.shape[0]):
#     fig.add_trace(go.Scatter3d(
#         x=[e_a[j, 0], e_i[j, 0], e_t[j, 0], e_a[j, 0]],  # Repeating e_a to close the loop
#         y=[e_a[j, 1], e_i[j, 1], e_t[j, 1], e_a[j, 1]],  # Repeat for y
#         z=[e_a[j, 2], e_i[j, 2], e_t[j, 2], e_a[j, 2]],  # And z
#         mode='lines',
#         line=dict(color='grey', width=1),
#         showlegend=False
#     ))

# Set labels and title
fig.update_layout(
    scene=dict(
        xaxis_title='Component 1',
        yaxis_title='Component 2',
        zaxis_title='Component 3'
    ),
    title=f'{loss_fn} UMAP, metric = {metric}'
)
fig.show()
```
