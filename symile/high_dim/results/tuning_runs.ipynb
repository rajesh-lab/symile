```python
df = df_10
loss_fn = "symile"
```

```python
# get rid of the second repeat epoch

# Step 1: Convert run_name to datetime
df['datetime'] = pd.to_datetime(df['run_name'].str[:15], format='%Y%m%d_%H%M%S')

# Step 2: Sort the DataFrame by the datetime column
df = df.sort_values(by='datetime')

# Step 3: Drop duplicates based on the specified columns, keep the first (earliest due to sort)
df = df.drop_duplicates(subset=['loss_fn', 'epoch', 'weight_decay', 'learning_rate', 'wandb_path'], keep='first')
```
